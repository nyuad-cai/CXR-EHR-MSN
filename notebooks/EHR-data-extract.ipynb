```python
import os
import numpy as np
import pandas as pd 
```

```python
mimic_iv_patients = 'local-path-where-mimic-located/physionet.org/files/mimic-iv-1.0/core/patients.csv'
mimic_iv_admissions = 'local-path-where-mimic-located/physionet.org/files/mimic-iv-1.0/core/admissions.csv'
mimic_iv_icu_stays = 'local-path-where-mimic-located/physionet.org/files/mimic-iv-1.0/icu/icustays.csv'

mimic_cxr_meta = 'local-path-where-mimic-located/physionet.org/files/mimic-cxr-jpg/2.0.0/mimic-cxr-2.0.0-metadata.csv'

patient_images = pd.read_csv(mimic_cxr_meta)
```

```python
#list(patient_images)
patient_da = pd.read_csv(mimic_iv_patients)
patient_da
```

```python
cxr_dataset = patient_images
```

```python
cols = ['dicom_id', 'subject_id', 'study_id', 'StudyDate', 'ViewPosition', 'PatientOrientationCodeSequence_CodeMeaning']
cxr_dataset = patient_images[cols]
cxr_dataset = cxr_dataset.rename(columns={"ViewPosition": "cxr_view"})
```

```python
cxr_dataset = cxr_dataset.rename(columns={"PatientOrientationCodeSequence_CodeMeaning": "cxr_orientation"})
```

```python
cxr_dataset
```

```python
patient_data = pd.read_csv(mimic_iv_patients)
```

```python
final_dataset = cxr_dataset.merge(patient_data[['subject_id', 'gender']], on='subject_id', how='left')
final_dataset
```

```python
patients_adm = pd.read_csv(mimic_iv_admissions)
new_patients_adm = patients_adm.drop_duplicates(subset=['subject_id'], keep='last')
```

```python
final_dataset_2 = final_dataset.merge(new_patients_adm[['subject_id', 'ethnicity', 'hadm_id']], on='subject_id', how='left')
final_dataset_2
```

```python
final_dataset_2['study_date'] = final_dataset_2['StudyDate'].apply(lambda x: int(str(x)[:4]))
final_dataset_2
```

```python
# mimic_iv_icu_stays
icu_stays_table = pd.read_csv(mimic_iv_icu_stays)
icu_stays_table
```

```python
final_dataset_2 = final_dataset_2.merge(patient_data[['subject_id', 'anchor_age', 'anchor_year']], on='subject_id', how='left')
final_dataset_2
```

```python
birth_year = final_dataset_2['anchor_year'] - final_dataset_2['anchor_age']
age = final_dataset_2['study_date'] - birth_year
```

```python
final_dataset_2['age'] = age
```

```python
final_dataset_2
```

```python
final_dataset_2['stay_expire'] = 0
from tqdm import tqdm
```

```python
# looping through users!
for index, row in tqdm(final_dataset_2.iterrows()):
    cxr_patient = row['subject_id']
    cxr_date = row['StudyDate']
    # check admission entries for patient
    cxr_admissions = patients_adm[patients_adm['subject_id'] == cxr_patient]
    for index2, row2 in cxr_admissions.iterrows():
        admittime = ''.join(row2['admittime'][:10].split('-'))
        dischtime = ''.join(row2['dischtime'][:10].split('-'))
        if cxr_date >= int(admittime) and cxr_date <= int(dischtime):
            #print(index, row['hospital_expire'])
            #print(index, row2['hospital_expire_flag'])
            final_dataset_2.at[index,'stay_expire'] = row2['hospital_expire_flag']
```

```python
final_dataset_2['stay_expire'].value_counts()
```

```python
final_dataset_2['icu_stay_los'] = 0
final_dataset_2['icu_stay'] = 0
```

```python
# looping through users!
for index, row in tqdm(final_dataset_2.iterrows()):
    cxr_patient = row['subject_id']
    cxr_date = row['StudyDate']
    # check admission entries for patient
    icu_admissions = icu_stays_table[icu_stays_table['subject_id'] == cxr_patient]
    for index2, row2 in icu_admissions.iterrows():
        intime = ''.join(row2['intime'][:10].split('-'))
        outtime = ''.join(row2['outtime'][:10].split('-'))
        if cxr_date >= int(intime) and cxr_date <= int(outtime):
            #print(index, row['hospital_expire'])
            #print(index, row2['hospital_expire_flag'])
            final_dataset_2.at[index,'icu_stay_los'] = row2['los']
            final_dataset_2.at[index,'icu_stay'] = 1
```
